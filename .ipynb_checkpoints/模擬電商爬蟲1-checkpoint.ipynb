```python
import requests
from pyquery import PyQuery as pq
```

```python
res = requests.get("http://webscraper.io/test-sites/e-commerce/allinone")
doc = pq(res.text)
```

```python
dataSet = []
doc.make_links_absolute(base_url=res.url)
for eachCateDoc in doc("#side-menu > li:nth-child(n+2) >\
a").items():
    lv1Doc = pq(eachCateDoc.attr("href"))
    lv1Doc.make_links_absolute()
    for eachSubCateDoc in lv1Doc("#side-menu > li.active > ul > \
    li > a").items():
        lv2Doc = pq(eachSubCateDoc.attr("href"))
        for eachItemDoc in lv2Doc(".thumbnail").items():
            itemDict = {}
            itemDict["title"]=eachItemDoc(".caption .title").text()
            itemDict["price"]=eachItemDoc(".caption .price").text()
            dataSet.append(itemDict)
            
```
